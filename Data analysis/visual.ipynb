```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(-10,10,0.1)
print(x.shape)
```

```python
x2=x**2
```

```python
plt.plot(x,x2)
```

```python
plt.plot(x, x2, 'go--', linewidth=2, markersize=12)
```

```python
x3=x**3
```

```python
plt.plot(x, 'go--', linewidth=2, markersize=12)
```

```python
plt.plot(x, 'go--', linewidth=2, markersize=5)
plt.plot(x2, 'go--', linewidth=2, markersize=5)
plt.plot(x3, 'go--', linewidth=2, markersize=5)
```

```python
plt.plot(x, label='original')
plt.plot(x2, label='squared')
plt.plot(x3, label='cubed')
plt.legend()
plt.show()
```

# Size of the graph

```python
plt.figure(figsize=(10,10))  # Size of the graph to be addresed
plt.plot(x, label='original')
plt.plot(x2, label='squared')
plt.plot(x3, label='cubed')
plt.legend()
plt.show()
```

# For the lines to be modified

```python
plt.figure(figsize=(10,10))  # Size of the graph to be addresed
plt.plot(x, label='original',linestyle=':')
plt.plot(x2, label='squared',linestyle='--')
plt.plot(x3, label='cubed')
plt.legend()
plt.show()
```

# Multiple entries
Alpha = Transperancy

```python
prices=[213,34,45,67,5,65,54,32]
qty=[12,43,53,46,56,45,23,34]
quality=[12,43,23,42,14,33,45,34]
plt.scatter(prices,qty,color='green',marker='x',s=100,alpha=.5)
plt.scatter(prices,quality,color='red',marker='o',s=10,alpha=.5,edgecolors='red')
```
